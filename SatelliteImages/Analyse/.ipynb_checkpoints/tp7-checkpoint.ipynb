## Méthode des K plus proches voisins 
Question 1 : 
- `n-neighbors` détermine le nombre de voisins 


```python
import pandas as pd 
import matplotlib.pyplot as plt
import utils

x  = pd.read_csv("Synth1-2000.csv",sep=",")
x0 = x.drop(columns=['y'])

utils.plot_clustering(data=x0,labels=x['y'])
```

Question 2 : 

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np 



cls = KNeighborsClassifier(n_neighbors=5)
cls.fit(x0,x['y'])
print("Les classes sont : ",cls.classes_,"\n")

Y = np.array([[0, 0], [0, -1], [1, 0], [1, 1]])
labels = cls.predict(Y)
print("On obtient comme classification pour Y:") 
print(labels,"soit P1 : 0 , P2 : 1 ; P3 : 0 et P4: 1")


plt.scatter(*Y.T, color="k")
```
